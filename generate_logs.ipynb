```python
import random
from datetime import datetime, timedelta
import pandas as pd
```

```python
columns = [
    "timestamp",
    "ip_address",
    "endpoint",
    "status_code",
    "response_time_ms"
]
```

```python
endpoints = [
    "/login",
    "/logout",
    "/dashboard",
    "/api/data",
    "/api/report"
]
status_codes = [200, 200, 200, 200, 404, 500]
```

```python
data = []

start_time = datetime.now()

for i in range(1000):
    log = {
        "timestamp": start_time + timedelta(seconds=i),
        "ip_address": f"192.168.1.{random.randint(1, 254)}",
        "endpoint": random.choice(endpoints),
        "status_code": random.choice(status_codes),
        "response_time_ms": random.randint(50, 1500)
    }
    data.append(log)

df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
df.to_csv("server_logs.csv", index=False)
```
